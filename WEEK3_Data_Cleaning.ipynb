```python
df=df.drop_duplicates()
df.shape
```

```python
df['Rating'] = df['Rating'].fillna('Unknown')
df.isna().sum()
```

```python
avg = df['Age'].mean()
df['Age'] = df['Age'].fillna(avg)
df.isna().sum()
```

```python
df.describe()
```

```python
df['Established'].value_counts()
```

```python
a = df[df['Established'] != -1]
b=a['Established'].mean()
b
```

```python
df['Established']=df['Established'].replace(-1,b)
```

```python
df.head(10)
```

```python
df['Easy Apply'].value_counts()
```

```python
df['Easy Apply']=df['Easy Apply'].replace("-1","FALSE")
df.head(10)
```

```python
df['Age'] = df['Age'].astype(int)
df.head(10)
```

```python
df['Established'] = df['Established'].astype(int)
df.head(10)
```

```python
df['Salary']=df['Salary'].apply(lambda x:x.replace("$",""))
df['Salary']=df['Salary'].apply(lambda x:x.replace("k","000"))
df['Salary']
```

```python
df["Location"]=df['Location'].str.replace('Australia Aus','Australia,Aus')
```

```python
df["Location"]=df['Location'].str.replace('India In','India,In')
```
