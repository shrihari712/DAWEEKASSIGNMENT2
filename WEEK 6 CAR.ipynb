```python
df.loc[df['Fuel_Type'] != 'CNG', 'ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].fillna(df['Highway_Mileage'])
```

```python
df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].str.replace(r'\s*km/litre', '', regex=True)
df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].str.replace('9.8-10.0', '10.0')
df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].str.replace('22.4-21.9', '22.4')
df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].str.replace(r'\s*kmpl', '', regex=True)

df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].astype(float)

df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].replace(142.0, 14.2)
df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].replace(1449.0, 14.49)

df.rename(columns={'ARAI_Certified_Mileage': 'ARAI_Certified_Mileage_kmpl'}, inplace=True)
```

```python
df['ARAI_Certified_Mileage_for_CNG'] = df['ARAI_Certified_Mileage_for_CNG'].str.replace(r'\s*km/kg', '', regex=True)
df['ARAI_Certified_Mileage_for_CNG'] = df['ARAI_Certified_Mileage_for_CNG'].astype(float)

df.rename(columns={'ARAI_Certified_Mileage_for_CNG': 'ARAI_Certified_Mileage_for_CNG_kmpkg'}, inplace=True)
```

```python
df['ARAI_Certified_Mileage_kmpl'].fillna(df['ARAI_Certified_Mileage_for_CNG_kmpkg'], inplace=True)
```

```python
mileage_drop = ['City_Mileage', 'Highway_Mileage','ARAI_Certified_Mileage_for_CNG_kmpkg']
df.drop(columns=mileage_drop, inplace=True)
```

```python
df.loc[df['Make'].str.contains('Mercedes-Benz', na=False), 'Make'] = 'Mercedes-Benz'
df.loc[df['Make'].str.contains('Rolls-Royce', na=False), 'Make'] = 'Rolls-Royce'
df.loc[df['Make'].str.contains('Go+', na=False), 'Make'] = 'Datsun'
df.loc[df['Make'].str.contains('Maruti Suzuki R', na=False), 'Make'] = 'Maruti Suzuki'
df['Make'] = df['Make'].str.replace('Land Rover Rover', 'Land Rover')
```

```python
df.loc[df['Make'].str.contains('Wagon', na=False), 'Make'] = 'Wagon R'
df['Model'] = df['Model'].str.replace('Wagon', 'Wagon R')
df['Model'] = df['Model'].str.replace('Mercedes-Benz ', '')
df['Model'] = df['Model'].str.replace('Rolls-Royce ', '')
df['Variant'] = df['Variant'].str.replace('Datsun ', '')
```

```python
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].replace(r'\D', '', regex=True)
```

```python
df['Displacement'] = df['Displacement'].replace(r'\D', '', regex=True)

df.rename(columns={'Displacement': 'Displacement_cc'}, inplace=True)

df['Displacement_cc'] = df['Displacement_cc'].astype(float)
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns

outliers = {}
for column in numeric_columns:
    # Filter values where they exceed the mean plus 3 times the standard deviation
    column_outliers = df.loc[df[column] > df[column].mean() + 3 * df[column].std(), column].values
    outliers[column] = column_outliers

# Print values of outliers for each numeric column
for column, values in outliers.items():
    print(f"Values of Outliers in {column}:\n", values)
```

```python
average_price_by_make = df.groupby('Make')['Ex-Showroom_Price_INR'].mean()
print("Average Ex-showroom Price by Make:\n", average_price_by_make)
```

```python
super_expensive_range = (df['Ex-Showroom_Price_INR'] > 5000000)
expensive_range = ((df['Ex-Showroom_Price_INR'] <= 5000000) & (df['Ex-Showroom_Price_INR'] > 2000000))
affordable_range = (df['Ex-Showroom_Price_INR'] <= 2000000)

df['price_segment'] = pd.cut(df['Ex-Showroom_Price_INR'], bins=[0, 2000000, 5000000, float('inf')],
                              labels=['Affordable', 'Expensive', 'Super Expensive'], right=False)

df.groupby(['price_segment', 'Make'])['Ex-Showroom_Price_INR'].mean().unstack().transpose()
```

```python
import seaborn as sns                    
import plotly.express as px 
```

```python
fig, axes = plt.subplots(1,2, figsize=(14, 4))

log_bins = np.logspace(np.log10(df['Ex-Showroom_Price_INR'].min()), np.log10(df['Ex-Showroom_Price_INR'].max()), 20)
axes[0].hist(df['Ex-Showroom_Price_INR'], bins=log_bins, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Ex-showroom Prices')
axes[0].set_xlabel('Ex-showroom Price (INR)')
axes[0].set_ylabel('Frequency')
axes[0].set_xscale('log')  

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))


sns.countplot(data=df, x='Make', palette='viridis', order=df['Make'].value_counts().index, ax=axes[0])
axes[0].set_title('Count of Diversity of Car Makes')
axes[0].set_xlabel('Car Make')
axes[0].set_ylabel('Total Count they Offer')
axes[0].tick_params(axis='x', rotation=90)


top_models_by_make = df.groupby('Make')['Model'].apply(lambda x: x.value_counts().idxmax())


top_diverse_cars = df[df['Model'].isin(top_models_by_make)]


sns.countplot(data=top_diverse_cars, x='Model', order=top_diverse_cars['Model'].value_counts().index, palette='viridis', ax=axes[1])
axes[1].set_title('Top-Diverse Model from Each Make')
axes[1].set_xlabel('Car Model')
axes[1].set_ylabel('Counts of Variants')
axes[1].tick_params(axis='x', rotation=90)


plt.tight_layout()
plt.show()
```
